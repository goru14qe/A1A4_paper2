```python
# Clear all variables
%reset -f
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
from scipy.interpolate import griddata
from matplotlib.colors import Normalize
from matplotlib import cm
import re
import glob
```

```python
input_folder = '../irregular/sim/irregular_0.07_extended/pos1'
output_folder = '../irregular/sim/irregular_0.07_extended/Pos1_transformed'
filename = 'SIM_L17_Pos01_Re100_B0001_T.csv'
```

```python
x = 'Points_0'
y = 'Points_1'
z = 'Points_2'
Vel_u = 'av_u_0'
Vel_v = 'av_u_1'
Vel_w = 'av_u_2'
# x = 'x'
# y = 'y'
# z = 'z'
# Vel_u = 'Vel_u'
# Vel_v = 'Vel_v'
# Vel_w = 'Vel_w'
```

```python
#Irregular degree test
input_file_path = os.path.join(input_folder, filename)
name, ext = os.path.splitext(filename)

# Extract layer (L) and position (Pos) information from the filename
layer_match = re.search(r'L(\d{2})', name)
pos_match = re.search(r'Pos(\d{2})', name)

if layer_match and pos_match:
    layer = int(layer_match.group(1))
    pos = int(pos_match.group(1))
    print(f"Layer: {layer}, Position: {pos}")
else:
    print(f"Filename does not contain valid layer or position information: {filename}")

# Since we're working with a CSV file, load it directly with pandas
try:
    df = pd.read_csv(input_file_path)
    print(f"Successfully loaded CSV with {df.shape[0]} rows and {df.shape[1]} columns")
    print(f"Columns: {df.columns.tolist()}")
except Exception as e:
    print(f"Error loading file: {str(e)}")

# FOR IRREGULAR ANGLES (currently active)
if 'L17' in name or 'L_17' in name:
    alpha = -5 * np.pi / 9        # -100°

elif 'L16' in name or 'L_16' in name:
    alpha = -31 * np.pi / 36      # -155°

elif 'L15' in name or 'L_15' in name:
    alpha = -31 * np.pi / 18      # -310°

elif 'L14' in name or 'L_14' in name:
    alpha = -7 * np.pi / 4        # -315°

elif 'L13' in name or 'L_13' in name:
    alpha = -9 * np.pi / 4        # -405°

elif 'L12' in name or 'L_12' in name:
    alpha = -103 * np.pi / 36     # -515°

elif 'L11' in name or 'L_11' in name:
    alpha = -29 * np.pi / 9       # -580°

elif 'L10' in name or 'L_10' in name:
    alpha = -131 * np.pi / 36     # -655°

elif 'L09' in name or 'L_09' in name:
    alpha = -19 * np.pi / 3       # -760°

elif 'L08' in name or 'L_08' in name:
    alpha = -89 * np.pi / 18      # -890°

elif 'L07' in name or 'L_07' in name:
    alpha = -187 * np.pi / 36     # -935°

elif 'L06' in name or 'L_06' in name:
    alpha = -11 * np.pi / 2       # -990°

elif 'L05' in name or 'L_05' in name:
    alpha = -215 * np.pi / 36     # -1075°

elif 'L04' in name or 'L_04' in name:
    alpha = -109 * np.pi / 36     # -1090°

elif 'L03' in name or 'L_03' in name:
    alpha = -20 * np.pi / 3       # -1200°

elif 'L02' in name or 'L_02' in name:
    alpha = -131 * np.pi / 18     # -1310°
  
else:
    alpha = 0
    print(f"Warning: Layer not recognized, using alpha = 0")

# Determine position parameter 'a'
if 'Pos_01' in name or 'Pos01' in name:
    a = 22.5
elif 'Pos_03' in name or 'Pos03' in name:
    a = 7.5
elif 'Pos_02' in name or 'Pos02' in name:
    a = 15
elif 'Pos_04' in name or 'Pos04' in name:
    a = 0
elif 'Pos_05' in name or 'Pos05' in name:
    a = -7.5
elif 'Pos_06' in name or 'Pos06' in name:
    a = -15
else:  # Assume Pos_07
    a = -22.5
    
print (f"Alpha: {alpha}, a: {a}")
print(f"Alpha (degrees): {np.degrees(alpha):.3f}, a (degrees): {a}")
print(f'a: {a}')

# === TRANFORMATION ===
file_path = os.path.join(input_folder, filename)
df = pd.read_csv(file_path)
df_transformed = df.copy()

df_transformed[x] = df[x]*np.cos(alpha) + df[y]*np.sin(alpha)
df_transformed[y] = -df[x]*np.sin(alpha) + df[y]*np.cos(alpha)
df_transformed[z] = df[z]
df_transformed[Vel_u] = df[Vel_u] * np.cos(alpha) + df[Vel_v]*np.sin(alpha)
df_transformed[Vel_v] = df[Vel_u] * np.sin(alpha) + df[Vel_v]*np.cos(alpha)
df_transformed[Vel_w] = df[Vel_w]
df_transformed['Vel_mag'] = np.sqrt(df_transformed[Vel_u]**2 + df_transformed[Vel_w]**2)

out_filename = f"{name}_transformed{ext}"
out_path = os.path.join(output_folder, out_filename)
df_transformed.to_csv(out_path, index=False)
print(f"Saved transformed file to: {out_path}")
```
