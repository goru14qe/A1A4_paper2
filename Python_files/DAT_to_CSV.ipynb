```python
# Clear all variables
%reset -f
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
from scipy.interpolate import griddata
from matplotlib.colors import Normalize
from matplotlib import cm
import re
import glob
```

```python
dat_folder = '../30/Exp_Data/Re300/'
out_folder_dat2csv = os.path.join(dat_folder, 'converted_csv')
os.makedirs(out_folder_dat2csv, exist_ok=True)

dat_files = glob.glob(os.path.join(dat_folder, '*.dat'))

for dat_file in dat_files:
    base = os.path.splitext(os.path.basename(dat_file))[0]
    out_path = os.path.join(out_folder_dat2csv, base + '.csv')
    
    # Always read using whitespace separator
    df_conv = pd.read_csv(
        dat_file,
        skiprows=3,
        sep=r'\s+',     # <-- splits on ANY amount of whitespace
        engine='python',
        header=None
    )
    
    # Save as true CSV (comma separated)
    df_conv.to_csv(out_path, index=False, sep=',')   # semicolon = Excel friendly
    
    print(f'Converted: {dat_file} -> {out_path}')
```

```python
# replace the first row (header) of all csv files in out_folder_dat2csv
csv_files = glob.glob(os.path.join(out_folder_dat2csv, '*.csv'))
new_cols = ['x','y','z','Vel_u','Vel_v','Vel_w','Vel_mag',
            'Stand_dev_Vx','Stand_dev_Vy','Stand_dev_Vz','Numb_of_vectors','isValid']

for csv_file in csv_files:
    with open(csv_file, 'r', newline='') as f:
        lines = f.readlines()
    # detect delimiter from existing header (fallback to comma)
    if lines:
        detected_sep = '\t' if '\t' in lines[0] else ','
    else:
        detected_sep = ','
    header_line = detected_sep.join(new_cols) + '\n'
    new_lines = [header_line] + (lines[1:] if len(lines) > 1 else [])
    with open(csv_file, 'w', newline='') as f:
        f.writelines(new_lines)
    print(f'Updated header: {csv_file}')
```
